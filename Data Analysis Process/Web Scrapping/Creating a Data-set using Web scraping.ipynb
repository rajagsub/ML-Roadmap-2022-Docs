## Web Scraping | Creating a Data-set using Web scraping

```python
# Importing Library
import requests
import pandas as pd
from bs4 import BeautifulSoup 
import numpy as np
```

```python
# Adding fake user agent to access the webpage 
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
```

```python
webpage = requests.get('https://www.ambitionbox.com/list-of-companies?page=1', headers=headers).text
```

```python
soup = BeautifulSoup(webpage,'lxml')
```

```python
company = soup.find_all('div',class_='company-content-wrapper')
```

```python
name=[]
rating = [] 
reviews = []
ctype =[]
hq = []
old = []
employees = []

for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('p',class_='rating').text.strip())
    reviews.append(i.find('a',class_='review-count').text.strip())
    ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())
    hq.append(i.find_all('p',class_='infoEntity')[1].text.strip())
    old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
    try:
        employees.append(i.find_all('p',class_='infoEntity')[3].text.strip())
    except:
        employees.append(np.nan)

D={'name': name, 'rating':rating, 'reviews':reviews, 'type':ctype, 'hq':hq, 'old':old , 'employees': employees}

df = pd.DataFrame(D)
```

```python
final = pd.DataFrame()
for j in range(1,333):
    url = 'https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
    
    # Adding fake user agent to access the webpage 
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    webpage = requests.get(url, headers=headers).text
    
    soup = BeautifulSoup(webpage,'lxml')
    company = soup.find_all('div',class_='company-content-wrapper')
    name=[]
    rating = [] 
    reviews = []
    ctype =[]
    hq = []
    old = []
    employees = []

    for i in company:
        name.append(i.find('h2').text.strip())
        rating.append(i.find('p',class_='rating').text.strip())
        reviews.append(i.find('a',class_='review-count').text.strip())
        try:
            ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())
        except:
            ctype.append(np.nan)
        try:
            hq.append(i.find_all('p',class_='infoEntity')[1].text.strip())
        except:
            hq.append(np.nan)
        try:
            old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
        except: 
            old.append(np.nan)
        try:
            employees.append(i.find_all('p',class_='infoEntity')[3].text.strip())
        except:
            employees.append(np.nan)

    D={'name': name, 'rating':rating, 'reviews':reviews, 'type':ctype, 'hq':hq, 'old':old , 'employees': employees}

    df = pd.DataFrame(D)

    final = final.append(df,ignore_index=True)
    
```
