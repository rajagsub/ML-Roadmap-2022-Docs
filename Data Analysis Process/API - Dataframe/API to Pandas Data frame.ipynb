```python
# Sample code to write the data into Dataframe on single PAGE.

popularity    = []
vote_count    = []
original_lang = []
title         = [] 
vote_average  = []
overview      = []
release_date  = []

for i in response['results']:
    popularity.append(i['popularity'])
    vote_count.append(i['vote_count'])
    original_lang.append(i['original_language'])
    title.append(i['title'])
    vote_average.append(i['vote_average'])
    overview.append(i['overview'])
    release_date.append(i['release_date'])
    
# Creating the dictonary for the above information to append to dataframe
d = {
    'title' : title,
    'overview' : overview,
    'original_lang' :original_lang,
    'release_date' : release_date,
    'popularity': popularity,
    'vote_count': vote_count,
    'vote_average' : vote_average
    }
    
df = pd.DataFrame(d)
df
```

```python
# Complete code to loop over the ranges till 500 pages and write the below details into the dataframe and further then 
# write them to the CSV file

final = pd.DataFrame()

for j in range(1,501):
    response = requests.get('https://api.themoviedb.org/3/movie/popular?api_key=9367294d13adfda853eceaa9b05a5c68&language=en-US&page={}'.format(j)).json()
    
    popularity    = []
    vote_count    = []
    original_lang = []
    title         = [] 
    vote_average  = []
    overview      = []
    release_date  = []
    
    print('Page No:', j)
    
    try:
        for i in response['results']:
            popularity.append(i['popularity'])
            vote_count.append(i['vote_count'])
            original_lang.append(i['original_language'])
            title.append(i['title'])
            vote_average.append(i['vote_average'])
            overview.append(i['overview'])
            release_date.append(i['release_date'])
            
    except KeyError:
        print('Key Error')
        release_date.append(' ')
    
    finally:
        print("The 'try except' is finished")
         

    # Creating the dictonary for the above information to append to dataframe
    d = {
        'title' : title,
        'overview' : overview,
        'original_lang' :original_lang,
        'release_date' : release_date,
        'popularity': popularity,
        'vote_count': vote_count,
        'vote_average' : vote_average
        }

    df = pd.DataFrame(d)
    
    final = final.append(df, ignore_index=True)
    
    
```
